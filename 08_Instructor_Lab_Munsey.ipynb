```python
# Import necessary libraries
import pandas as pd

# Read in the dataset
df = pd.read_csv(r"C:\Users\TechnellogicPC\UTW\Python\week7labs\github_teams.csv")

# 1. Select the columns Team_type, human_work, and work_per_human
selected_cols_1 = (
    df
    .loc[:, ['Team_type', 'human_work', 'work_per_human']]
)
print("1. Selected Columns:")
print(selected_cols_1.head(), '\n')

# 2. Select columns that end in the letter t using regex t$
selected_cols_2 = (
    df
    .filter(regex='t$')
)
print("2. Columns ending in 't':")
print(selected_cols_2.head(), '\n')

# 3. Sort data descending using Team_size_class, human_work, work_per_human
df_sorted = (
    df
    .assign(Team_size_class=pd.Categorical(df['Team_size_class'], categories=['Small', 'Medium', 'Large'], ordered=True))
    .sort_values(by=['Team_size_class', 'human_work', 'work_per_human'], ascending=[False, False, False])
)
print("3. Sorted Data:")
print(df_sorted[['Team_size_class', 'human_work', 'work_per_human']].head(), '\n')

# 4. Select human-bot teams with bot_members_count >= 3
human_bot_large = (
    df
    .query("Team_type == 'human-bot' and bot_members_count >= 3")
)
print("4. Human-bot teams with 3+ bots:", human_bot_large.shape[0], "teams\n")

# 5. Find human teams that are Large with human_gini >= 0.75
human_large_inequality = (
    df
    .query("Team_type == 'human' and Team_size_class == 'Large' and human_gini >= 0.75")
)
print("5. Human Large teams with Gini >= 0.75:", human_large_inequality.shape[0], "teams\n")

# 6. How many teams are in the Small or Large category?
small_or_large_count = (
    df
    .query("Team_size_class in ['Small', 'Large']")
    .shape[0]
)
print("6. Small or Large teams count:", small_or_large_count, '\n')

# 7. How many teams are in the Small or Large category with a human_gini <= 0.25?
low_inequality_small_or_large = (
    df
    .query("Team_size_class in ['Small', 'Large'] and human_gini <= 0.25")
    .shape[0]
)
print("7. Small or Large teams with Gini <= 0.25:", low_inequality_small_or_large, '\n')

# 8. How many human teams are in the Medium category?
medium_human_count = (
    df
    .query("Team_type == 'human' and Team_size_class == 'Medium'")
    .shape[0]
)
print("8. Medium human-only teams count:", medium_human_count, '\n')

# 9. Save the columns Team_size_class and work_per_human as a new DataFrame
team_size_and_work = (
    df
    .loc[:, ['Team_size_class', 'work_per_human']]
)
print("9. New DataFrame with Team_size_class and work_per_human:")
print(team_size_and_work.head(), '\n')

# 10. Rename columns: human_gini → work_inequality and eval_survival_day_median → issue_resolution_time
renamed_df = (
    df
    .rename(columns={
        'human_gini': 'work_inequality',
        'eval_survival_day_median': 'issue_resolution_time'
    })
)
print("10. Renamed columns:")
print(renamed_df[['work_inequality', 'issue_resolution_time']].head())
```
